```python
word = "orange"
print(word[-5:-2])
print(word[::-1])  # tersinde yazdırma
print(word[-2:-5:-1])
```

```python
print(word[4:1])  # boş string verir!!!  Slice'lama işlemi soldan sağa doğru (step değeri default +1) yapılır. 
print(word[-3:-6])  # boş string verir!!! Slice'lama işlemi soldan sağa doğru (step değeri default +1) yapılır.
print(word[4:1:-1])  # step -1 yapılarak çıktı alındı...
print(word[-3:-6:-1])  # step -1 yapılarak çıktı alındı...
```
