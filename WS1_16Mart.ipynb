**Problem 2:**
Write a program that calculates body mass index from height and weight entered by the user.

Body mass index : Weight / Height(m) x Height(m)

```
height = 1.75
weight = 85
body_mass_index = weight / (height * height)  # Body mass index : Weight / (Height(m) x Height(m))
print("Your body mass index is:", body_mass_index)
```

**Problem 3:**
With your 200 usdollars, how many pieces of material can you get for $ 11 each? How much money do you have left after buying?

```
money = 200
unit_price = 11
num_of_units = money // unit_price
remaining_money = money %  unit_price
print("Left money is $", remaining_money, sep = "")
```

**Problem 6:**
Get word, seperator and repetition values from the user and print the word as many as repetition value on the screen according to the given values.

Ex: word = ali, sep = /, repetition = 3 ----> ali/ali/ali

```
word = input("Enter a word: ")
sep_i = input("Enter a seperator like /, #, * etc. :")
rep = int(input("Enter a repetition value :")) # “TypeError: can’t multiply sequence by non-int of type ‘str'” hatasından
output = rep * (word + sep_i)                   # kurtulmak için int olarak atandı.
print(output + "\b")                              
```
