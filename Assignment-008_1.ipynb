Problem : If you had deposited a coin on the cryptocurrency exchange that brought 7% fixed profit daily for a week, how much would your $ 1000 reach at the end of the 7th day?

Not: Ödevlerinizi gördüğünüz konular dahilindeki bilgileriniz ile yapınınız.

Bileşik Faiz formülü= a*(1+t/100)^n      t=faiz oranı n=dönem tekrarı

```
deposit = 1000
profit_rate = 7
num_of_days = 7
total_profit = deposit * ((1 + (profit_rate / 100) ) ** num_of_days)
print("Total money including profit of 1000$ after 7 days with 7% fixed daily profit will be:", total_profit)
```
