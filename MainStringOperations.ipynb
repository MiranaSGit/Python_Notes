**#### Searching a String ####**

```python
#### Searching a String ####
text = 'www.clarusway.com'
print(text.endswith('.com'))
print(text.startswith('http:'))

print(text.endswith('om'))
print(text.startswith('w'))


email = "clarusway@clarusway.com is my e-mail address"
print(email.startswith("@", 9))
print(email.endswith("-", 10, 32))
```

**### Changing a String ###**

```python
### Changing a String ###

print("Actions speaks louder than words".upper().swapcase().capitalize())
print("Actions speaks louder than words".upper())
print("Actions speaks louder than words".swapcase())
print("Actions speaks louder than words".capitalize())


city = "SARAJEVO"
text = f"I live in {city.capitalize()}"
print(text)

var_1 = 'Clarusway'
print(var_1.upper())
print(var_1)
var_1 = 'Clarusway'.upper()
print(var_1)


name = "MARIAM"
print(f"My name is {name.capitalize()}")


text = "clarusway@clarusway.com is my e-MAIL address"
print(text.startswith("cl"))
print(text.endswith("cl", 0, 2))
print(text.replace("clarusway", "HR", 1))
print(text.swapcase())
print(text.title())
print(text.capitalize())


sentence = "I live in Ankara"
titled_sentence = sentence.title()
changed_sentence = sentence.replace("Ankara", "Ä°stanbul")
mixed_sentence = changed_sentence.swapcase().upper()
print(sentence, titled_sentence, changed_sentence, mixed_sentence, sep = "\n")
```

**#### Editing a String  ######**

```python
#### Editing a String  ######

space_string = "     listen first      "
print(space_string.strip())  # removes all spaces from both sides

source_string = "interoperability"
print(source_string.strip("yi"))  
# removes trailing "y" or "i" or "yi" or "iy" from both sides

print(source_string.lstrip("in"))  
# removes "i" or "n" or "in" or "ni" from the left side

space_string = "     listen first      "
print(space_string.rstrip())  # removes spaces from the right side

source_string = "interoperability"
print(source_string.rstrip("yt"))  
# removes "y" or "t" or "yt" or "ty" from the right side


print(source_string.split("e"))   ## Split function


txt = "hello, my name is Peter, I am 26 years old"
x = txt.split(", ")
print(x)
```
