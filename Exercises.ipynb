```
sum = 0
for i in range(10):
    if i == 0:
        sum = i
        print("Current Number: {0} Previous Number: {0}  Sum: {1} ".format(i, sum))
    else:
        sum = i
        sum = sum + (i-1)
        print("Current Number: {} Previous Number: {}  Sum: {} ".format(i, i-1, sum))
        
```

Body mass index : Weight / Height(m) x Height(m)

```
weight = int(input("Agırlıgınızı girin lütfen: "))
height = float(input("Boyunuzu girin lütfen: "))
print(weight / (height * height))
```

Problem 3: With your 200 usdollars, how many pieces of material can you get for $ 11 each? How much money do you have left after buying?

```
total_money = 200
unit_price = 11
num_of_pieces = total_money // unit_price
remaining_money = total_money % num_of_pieces
print("Alınan ürün miktarı:", num_of_pieces, "\nkalan para:", remaining_money, "\b$" )
```
